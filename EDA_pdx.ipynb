# Mutation? Only one sample, maybe not so good.

```python
tmp = pd.read_csv('pdx/mut_and_cn2.csv')
tmp
```

```python
tmp.Category.unique()
```

- **MutNovel**: A new mutation that has not been previously reported.
- **Del0.8**: A reduction in copy number (deletion), indicating a decrease from the normal copy number of 2 to 0.8.
- **MutKnownFunctional**: A mutation that has been reported before and confirmed to have functional effects.
- **Amp5**: An increase in copy number (amplification), indicating an increase in copy number up to 5.
- **MutLikelyFunctional**: A mutation believed to have functional effects, although it has not yet been fully confirmed.

```python
tmp.pivot_table(index='Gene', columns='Sample', values='Category', aggfunc='first')
```
